```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('Report.csv')
```

```python
df
```

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
```

```python
df
```

```python
df.fillna(0, inplace=True)
```

```python
df
```

### Microsoft: Experienced declines in revenue and net income in 2022 but rebounded in 2023.ar.

### Tesla: Saw significant drops in revenue and net income in 2022, with no improvement in 2023.

### Apple: Experienced slight fluctuations but generally maintained stable financial metrics, with a slight increase in revenue and net income in 2022 and stable performance in 2023.

### Overall, Microsoft and Apple showed resilience and recovery after declines in 2022, while Tesla struggled to improve its financial performance in 2023 after a sharp decline in the previous year.

## Calculating Year-by-Year growth rates for Total Assets, Total Liabilities and Cash flow from Operations Activities

```python
df['Assets Growth (%)'] = df.groupby('Company')['Total Assets'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby('Company')['Total Liabilities'].pct_change() * 100
df['Cash Flow from Operations Growth(%)'] = df.groupby('Company')['Cash Flow from Operating Activities'].pct_change() * 100
```

```python
df
```

```python
df.fillna(0, inplace=True)
df
```

```python
df.to_csv('final_report.csv')
```

```python
summary = df.groupby('Company').agg({
    'Revenue Growth (%)': 'mean',
    'Net Income Growth (%)': 'mean',
    'Assets Growth (%)' : 'mean',
    'Liabilities Growth (%)' : 'mean',
    'Cash Flow from Operations Growth(%)' :'mean'    
}).reset_index()
```

```python
print('Year-By-Year Average Growth Rates(%) :-')
summary
```

```python
summary.to_csv('summary_final_report.csv')
```
